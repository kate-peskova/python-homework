```python

# Importing the pathlib and csv library
from pathlib import Path
import csv


# Setting up the file path
csvpath = Path("02-Homework_02-Python_Instructions_PyBank_Resources_budget_data.csv")


# Calculating the sum of total profits/losses and the total number of periods

# Initializing variables

total_months = 0
net_total_profit_losses = 0


# Opening the csv file as an object
with open(csvpath, "r") as csvfile:

    # Passing in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter=",")

    # Reading the header row first
    csv_header = next(csvreader)
    # Print the header
    print(csv_header)

    # Iterating through the lines
    
    for row in csvreader:
        # Print the row
        print(row)

        # converting profit/loss data from string into integer

        profits_losses = int(row[1])


        # Summing up the profit/losses total and count the numbers in the text file
        
       
        net_total_profit_losses += profits_losses
        total_months += 1

     

# Converting csv file into a dictionary

data = {}
csvfile = open(csvpath)
for line in csvfile:
    line = line.strip('\n')
    (date, profits_losses) = line.split(",")
    data[date] = profits_losses



# Converting dictionary items into lists
# Separating a list of dates

data_list = list(data.keys())

# Modifying the list of dates by cutting off the header (to furhter match the number of lines in the profit/losses list)

new_data_list = data_list[1:]

# Converting profits/losses into list so we can further convert the data into integers
profits_losses_list = list(data.values())

# Cutting of the header (as it would not convert into int)
profits_losses_list_int = profits_losses_list[1:]

# Initializing a new list to use it as a container for profit/losses daily changes
changes = []

# Initializing variable to iterate throught profit/losses list
profits_losses_day_one = profits_losses_list[1]


# Creating a loop to fill the changes list with the daily profit/losses changes

for profits_losses in profits_losses_list_int:
    change = int(profits_losses) - int(profits_losses_day_one)
    changes.append(change)
    profits_losses_day_one = profits_losses
    
    


# Calculating the sum of daily changes to calculate the average 

total_changes = sum(changes)

# Calculating the average of daily changes

average_change = round(total_changes / (total_months - 1), 2)



# Creating new dictionary to combine both lists so we could match dates with the daily changes

dates_changes = {}
i = 0

for new_dictionary_date in new_data_list:
    dates_changes[new_dictionary_date] = changes[i]
    i += 1




# Initializing minimum key-value pair

greatest_decrease_date = ""
greatest_decrease_value = 0


# Initializing maximum key-value pair

greatest_increase_date = ""
greatest_increase_value = 0


# Defining greatest increase and decrease in profit/losses changes through the loop

for new_dictionary_date, changes in dates_changes.items():
    
    # Finding out the greatest decrease
    
    if greatest_decrease_value == 0:
        greatest_decrease_value = changes
        greatest_decrease_date = new_dictionary_date
    elif changes < greatest_decrease_value:
        greatest_decrease_value = changes
        greatest_decrease_date = new_dictionary_date

    # Finding out the greatest increase
    
    if changes > greatest_increase_value:
        greatest_increase_value = changes
        greatest_increase_date = new_dictionary_date

        


# Setting output file name

output_path = 'output.txt'


# Opening the output path as a file object

with open(output_path, 'w') as file:


    file.write("Financial Analysis\n")
    file.write("----------------------------\n")
    file.write(f"Total Months: {total_months}\n")
    file.write(f"Total: {net_total_profit_losses}\n")
    file.write(f"Average  Change: {average_change}\n")
    file.write(f"Greatest Increase in Profits: {greatest_increase_date} ({greatest_increase_value})\n")
    file.write(f"Greatest Decrease in Profits: {greatest_decrease_date} ({greatest_decrease_value})")

```
